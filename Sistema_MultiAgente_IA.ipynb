```python
# =====================================
# PROYECTO IA - SISTEMA MULTIAGENTE
# CELDA 1 - LIBRERIAS
# =====================================

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

from sklearn.impute import SimpleImputer

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

print("Librerías cargadas correctamente")
```

```python
# =====================================
# AGENTE 1 - NORMALIZADOR
# CARGA DEL DATASET
# =====================================

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

print("Dataset cargado correctamente")
print()

print("Dimensiones del dataset:")
print(df.shape)

print()
print("Primeras 5 filas:")
display(df.head())
```

```python
# =====================================
# AGENTE 1 - ANALISIS INICIAL
# =====================================

print("Información general")
print()

df.info()

print()
print("Valores nulos por columna")
print()

print(df.isnull().sum())
```

```python
# =====================================
# AGENTE 1 - LIMPIEZA DE COLUMNAS
# =====================================

df_limpio = df.copy()

columnas_eliminar = [
    "PassengerId",
    "Name",
    "Ticket",
    "Cabin"
]

df_limpio.drop(columns=columnas_eliminar, inplace=True)

print("Columnas eliminadas correctamente")
print()
print("Columnas restantes:")
print(df_limpio.columns.tolist())
```

```python
# =====================================
# AGENTE 1 - IMPUTACION
# =====================================

imputador_edad = SimpleImputer(strategy="median")
df_limpio["Age"] = imputador_edad.fit_transform(
    df_limpio[["Age"]]
)

imputador_embarked = SimpleImputer(strategy="most_frequent")
df_limpio["Embarked"] = imputador_embarked.fit_transform(
    df_limpio[["Embarked"]]
).ravel()

print("Valores faltantes corregidos")
print()

print(df_limpio.isnull().sum())
```

```python
# =====================================
# AGENTE 1 - CODIFICACION
# =====================================

encoder_sex = LabelEncoder()
encoder_embarked = LabelEncoder()

df_limpio["Sex"] = encoder_sex.fit_transform(
    df_limpio["Sex"]
)

df_limpio["Embarked"] = encoder_embarked.fit_transform(
    df_limpio["Embarked"]
)

print("Variables categóricas codificadas")
print()

display(df_limpio.head())
```

```python
# =====================================
# AGENTE 1 - ESCALADO
# =====================================

scaler = StandardScaler()

columnas_numericas = [
    "Age",
    "Fare"
]

df_limpio[columnas_numericas] = scaler.fit_transform(
    df_limpio[columnas_numericas]
)

print("Escalado completado")
print()

display(df_limpio.head())
```

```python
# =====================================
# AGENTE 1 - EXPORTACION
# =====================================

df_limpio.to_csv(
    "dataset_limpio.csv",
    index=False
)

print("Dataset limpio generado")
print()

print("Dimensiones:")
print(df_limpio.shape)
```

```python
# =====================================
# AGENTE 2 - CARGA DEL DATASET LIMPIO
# =====================================

df_entrenamiento = pd.read_csv(
    "dataset_limpio.csv"
)

print("Dataset limpio cargado")
print()

print(df_entrenamiento.shape)

display(df_entrenamiento.head())
```

```python
# =====================================
# AGENTE 2 - DEFINICION DE X E Y
# =====================================

X = df_entrenamiento.drop(
    columns=["Survived"]
)

y = df_entrenamiento["Survived"]

print("Variables predictoras:")
print(X.columns.tolist())

print()
print("Variable objetivo:")
print("Survived")
```

```python
# =====================================
# AGENTE 2 - TRAIN TEST SPLIT
# =====================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Entrenamiento:", X_train.shape)
print("Prueba:", X_test.shape)
```

```python
# =====================================
# AGENTE 2 - MODELO 1
# LOGISTIC REGRESSION
# =====================================

modelo_lr = LogisticRegression(
    max_iter=1000
)

modelo_lr.fit(
    X_train,
    y_train
)

pred_lr = modelo_lr.predict(
    X_test
)

print("Logistic Regression entrenado")
```

```python
# =====================================
# AGENTE 2 - MODELO 2
# DECISION TREE
# =====================================

modelo_dt = DecisionTreeClassifier(
    random_state=42
)

modelo_dt.fit(
    X_train,
    y_train
)

pred_dt = modelo_dt.predict(
    X_test
)

print("Decision Tree entrenado")
```

```python
# =====================================
# AGENTE 2 - MODELO 3
# RANDOM FOREST
# =====================================

modelo_rf = RandomForestClassifier(
    random_state=42
)

modelo_rf.fit(
    X_train,
    y_train
)

pred_rf = modelo_rf.predict(
    X_test
)

print("Random Forest entrenado")
```

```python
# =====================================
# AGENTE 2 - EVALUACION
# =====================================

resultados = pd.DataFrame({
    "Modelo": [
        "Logistic Regression",
        "Decision Tree",
        "Random Forest"
    ],
    "Accuracy": [
        accuracy_score(y_test, pred_lr),
        accuracy_score(y_test, pred_dt),
        accuracy_score(y_test, pred_rf)
    ],
    "Precision": [
        precision_score(y_test, pred_lr),
        precision_score(y_test, pred_dt),
        precision_score(y_test, pred_rf)
    ],
    "Recall": [
        recall_score(y_test, pred_lr),
        recall_score(y_test, pred_dt),
        recall_score(y_test, pred_rf)
    ],
    "F1": [
        f1_score(y_test, pred_lr),
        f1_score(y_test, pred_dt),
        f1_score(y_test, pred_rf)
    ]
})

display(resultados)
```
